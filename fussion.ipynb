```python
#!/usr/bin/env python3
import pandas as pd
import numpy as np
import numpy as np
```

```python
# Charger les deux datasets
zara = pd.read_csv("zara_data_clean.csv")
vinted = pd.read_csv("vinted_data_clean.csv")
```

```python
# Afficher les colonnes de chaque dataset pour vérifier les différences
print(zara.columns)
print(vinted.columns)
```

```python
zara = zara.rename(columns={"prix": "prix_zara"})
vinted = vinted.rename(columns={"prix": "prix_vinted"})
```

```python
# Fusionner les datasets en utilisant une jointure interne sur les colonnes "categorie" et "collection"
Dataset_final = pd.merge(
    vinted,
    zara,
    on=["categorie", "collection"],
    how="inner"
)
```

```python
print("Dataset_final.shape:", Dataset_final.shape)
Dataset_final.head()
```

```python
# Calculer la décote en pourcentage et l'ajouter comme une nouvelle colonne
Dataset_final["decote"] = (
    (Dataset_final["prix_zara"] - Dataset_final["prix_vinted"])
    / Dataset_final["prix_zara"]
) * 100
```
